```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])
```

```python
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
i = 0

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def calculate_model_output(w, b, x):
    m = x.shape # dimensions
    f = np.zeros(m) # return an array with the same dimensions filled with zeros
    for i in range(len(x)):
        f[i] = w * x[i] + b
    return f # f just an array of predictions to given x
```

```python
tmp_f = calculate_model_output(w, b, x_train)

plt.plot(x_train, tmp_f, c='b', label='Our Prediction')

plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

plt.title("Housing prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
