```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
data = pd.read_csv('data.csv')
x_train = data.iloc[:, 0]
y_train = data.iloc[:, 1]
plt.scatter(x_train, y_train)
plt.show()
```

```python
w = 0
b = 0

l = 0.0001
iterations = 1000

m = float(len(x_train))

#Gradient descent
for i in range(iterations):
    f = w*x_train + b
    dw = (1/m) * sum(x_train*(f - y_train))
    db = (1/m) * sum(f - y_train)
    w = w - l*dw
    b = b - l*db
    
print(w, b)
```

```python
f = w*x_train + b

plt.scatter(x_train, y_train)
plt.plot([min(x_train), max(x_train)], [min(f), max(f)], color='red')
plt.show()
```
